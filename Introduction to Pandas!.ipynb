## Manipulating Data

```python
# Convert all the strings in Lower case in a column

car_sales["Colour"] = car_sales["Colour"].str.lower()
car_sales
```

```python
# Concvert all the strings in Upper case in a column


car_sales["Colour"] = car_sales["Colour"].str.upper()
car_sales["Colour"]
```

```python
# Concvert all the strings in proper format in a column


car_sales["Colour"] = car_sales["Colour"].str.title()
car_sales["Colour"]
```
